## Bronze Extraction Notebook

The purporse of this notebook is to create bronze tables and extract data to it using Spark

###Read and Write Data

```python
sources = {
    "source_crm": ["cust_info", "prd_info", "sales_details"],
    "source_erp": ["CUST_AZ12", "LOC_A101", "PX_CAT_G1V2"]
}

for source, files in sources.items():
    for filename in files:
        df = (
            spark.read
            .option("header", "true")
            .option("inferSchema", "true")
            .csv(f"/Volumes/workspace/bronze/source_data/{source}/{filename}.csv")
        )
        df.write.mode("overwrite").saveAsTable(f"workspace.bronze.{source[-3:]}_{filename}")
```
